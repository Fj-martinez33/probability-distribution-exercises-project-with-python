# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

## Ejercicio 1

Escriba las funciones anteriores pero para distribuciones normales y uniformes.

```python
####IMPORTAMOS LIBRERIAS####

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, uniform

#Variables#
size = 0 #media
prob = 1 #desviación
```

```python
# Definimos todas las funciones de probabilidad relacionadas con la distribución normal#

#Calculamos la estimación puntual de la distribución normal#
def dnorm(x, size, prob=0.5):
    return norm.pdf(x = x, loc = size, scale = prob )

#Calculamos la acumulación de la distribución normal#
def pnorm(q, size, prob = 0.5):
    return norm.cdf(x = q, loc = size, scale = prob)

#Calculamos la función cuantil a partir de la distribución normal#
def qnorm(p, size, prob = 0.5):
    return norm.ppf(q = p, loc = size, scale = prob)

#Generar valores aletarios#
def rnorm(n, size, prob=0.5):
    return norm.rvs(loc = size, scale = prob, size=n)
```

```python
# Distribución normal(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probabilidad de que en una distribución normal(10, 0.2) tenga el valor 2: {dnorm(2, size = 10, prob = 0.2)}")
print(f"Probabilidad de que en una distribución normal(10, 0.2) se obtenga un valor menor a 2: {pnorm(2, size = 10, prob = 0.2)}")
print(f"Que valor en una distribución normal(10, 0.2) tiene una probabilidad acumulativa de 0.9? {qnorm(0.9, size = 10, prob = 0.2)}")
print(f"Genera 2000 valores aleatorios a partir de una distribución normal (10, 0.2): {rnorm(2000, size = 10, prob = 0.2)}")
```

```python
np.random.seed(42) #Como queremos que siempre sean los mismos valores marcamos la semilla procedural de los datos en 42. [El 42 es el sentido de la vida]

data = np.linspace(-5, 5, 2000)

pdf = dnorm(data, size, prob)
cdf = pnorm(data, size, prob)
ppf = qnorm(data, size, prob)

#Graficos
fig, axis = plt.subplots(figsize = (5,7), nrows= 3, ncols= 1)

axis[0].plot(data, pdf, "-r")
axis[1].plot(data, cdf, "-y")
axis[2].plot(data, ppf, "-g")
```

```python
# Definimos todas las funciones de probabilidad relacionadas con la distribución uniforme#

#Calculamos la estimación puntual de la distribución uniforme#
def duniform(x, size, prob=0.5):
    return uniform.pdf(x = x, loc = size, scale = prob )

#Calculamos la acumulación de la distribución uniforme#
def puniform(q, size, prob = 0.5):
    return uniform.cdf(x = q, loc = size, scale = prob)

#Calculamos la función cuantil a partir de la distribución uniforme#
def quniform(p, size, prob = 0.5):
    return uniform.ppf(q = p, loc = size, scale = prob)

#Generar valores aletarios#
def runiform(n, size, prob=0.5):
    return uniform.rvs(loc = size, scale = prob, size=n)
```

```python
import numpy as np

np.random.seed(42)

print(f"Probabilidad de que en una distribución uniforme(10, 0.2) tenga el valor 2: {duniform(2, size = 10, prob = 0.2)}")
print(f"Probabilidad de que en una distribución uniforme(10, 0.2) se obtenga un valor menor a 2: {puniform(2, size = 10, prob = 0.2)}")
print(f"Que valor en una distribución uniforme(10, 0.2) tiene una probabilidad acumulativa de 0.9? {quniform(0.9, size = 10, prob = 0.2)}")
print(f"Genera 2000 valores aleatorios a partir de una distribución uniforme (10, 0.2): {runiform(2000, size = 10, prob = 0.2)}")
```
